# **Modelo para Prediccion**

Se realiza un modelo para proyeccion usando lo que se va a determinar como:

**Variable Predictoras**: overall', 'shooting','passing', 'dribbling', 'defending

**Variable Target**: Potential

```python
df_proyeccion = jf_df[['overall', 'shooting','passing', 'dribbling', 'defending','potential']].copy()
df_proyeccion.head(10)
```

## **Division de Dataset Training y Test**

1) Separar los datos de entrenamiento y de Test, se dividiran en un 70% los datos de entrenamiento y 30% datos de Test.

```python
datos_training= df_proyeccion.sample(frac=0.7,random_state=0)
datos_test= df_proyeccion.drop(datos_training.index)
```

2) Retirar del nuevo data frame de proyeccion, la variable Target que es la que queremos predecir "Potential"

```python
e_training= datos_training.pop("potential")
e_test= datos_test.pop("potential")
```

```python
e_training.head(10)
```

```python
datos_training.head(10)
```

3) Aplicar un modelo basico de regresion Lineal

```python
from sklearn.linear_model import LinearRegression
modelo= LinearRegression()
modelo.fit(datos_training, e_training)
```

```python
prediccion= modelo.predict(datos_test)
prediccion
```

4) Evaluar Sobre un nuevo jugador con datos aleatorios

```python
Value_Players= pd.DataFrame(np.array([[70,78,58,65,70]]),columns=["overall", "shooting", "passing", "dribbling", "defending"])
print([Value_Players])                        
```

5) Segunda Forma Division de datos Train y Test

```python
Y_col = 'potential' # me quedo con la columna target 
X_cols = df_proyeccion.loc[:, df_proyeccion.columns != 'prediccion'].columns
```

```python
from sklearn.model_selection import train_test_split
DatosX_train, DatosX_test, Datosy_train, Datosy_test = train_test_split(df_proyeccion[X_cols], df_proyeccion[Y_col],test_size=0.2, random_state=42)
```

6) Evaluando Sobre datos Test

```python
DatosX_test.head()
```

```python
Datosy_test.head()
```

7) Definicion Modelo Arbol de Decisiones

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import confusion_matrix, classification_report
```

```python
modelo_arbol_desiciones = DecisionTreeRegressor(max_depth=15,
                                random_state=50,
                                criterion='squared_error')
```

8) Entrenamiento de modelo con los datos Train

```python
modelo_arbol_desiciones.fit(DatosX_train,Datosy_train)
```

9) Prediccion con Datos Train

```python
modelo_arbol_desiciones_pred = modelo_arbol_desiciones.predict(DatosX_test)
```

```python
modelo_arbol_desiciones_pred
```

10) Metricas Obtenidas para los Datos

10.1) Error absoluto medio

```python
from sklearn.metrics import mean_absolute_error
val_mae = mean_absolute_error(Datosy_test,modelo_arbol_desiciones_pred)
val_mae
```

10.2) Error cuadratico medio

```python
from sklearn.metrics import mean_squared_error
val_mse = mean_squared_error(Datosy_test,modelo_arbol_desiciones_pred)
val_mse
```
